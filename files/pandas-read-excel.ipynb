### Luftfeuchtigkeit als ganzzahlige Prozentzahl

```python
def float_to_percent(number):
    return int(number * 100)
```

```python
df = pd.read_excel(
    "not_normalized.xlsx",
    header=3,
    converters={
        "Eigenschaft1": is_equal_to_x,
        "Luftfeuchtigkeit": float_to_percent,
    },
    true_values=["Ja"],
    false_values=["Nein"],
    usecols=["time_stamp", "Ertrag", "Luftfeuchtigkeit", "Temperatur", "Eigenschaft1", "Eigenschaft1", "Fehlerquote"],
    parse_dates=["time_stamp"],
    index_col=0,
)
df
```
